```python
# so both dataframes are the same, all good
#df = pd.read_csv('results_good_eval_9_11_2024.csv')
df = pd.read_csv('results_good_eval_2_8_2025_morph2_effcientnetv2_best_2.45.csv')
```

```python
df['age_diff'] = df['age_preds'] - df['age_labels']
```

```python
print(f"CS(5):{np.sum(df['abs_age_diff'] <= 5) / len(df['abs_age_diff'])}")
```

```python
sns_plot = sns.histplot(
    data=df, 
    x="age_diff",
    bins=20,
    stat="probability", 
    multiple="dodge",
    common_norm=False,
    shrink=.8
)

sns_plot.set(xlabel='Mean Estimation Error [age]', ylabel='Probability')
sns_plot.set(ylim=(0, None))
sns_plot.set(xlim=(-15, 15))
#sns_plot.get_figure().savefig('./plots/estimation_error_not_abs.pdf',dpi=300)
plt.savefig("./plots/good_eval_efficientnet_3_8_2025/mean_error_hist.pdf", format="pdf", bbox_inches="tight") 
```

```python
sns_plot = sns.histplot(
    data=df, 
    x="abs_age_diff",
    bins=20,
    stat="probability", 
    multiple="dodge",
    common_norm=False,
    shrink=.8
)

sns_plot.set(xlabel='Mean Absolute Estimation Error [age]', ylabel='Probability')
sns_plot.set(ylim=(0, None))
sns_plot.set(xlim=(0, 15))
#sns_plot.get_figure().savefig('./plots/estimation_error.pdf',dpi=300)
plt.savefig("./plots/good_eval_efficientnet_3_8_2025/mae_hist.pdf", format="pdf", bbox_inches="tight") 
```
