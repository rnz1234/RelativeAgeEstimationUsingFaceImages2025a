```python
import numpy as np
import json
```

```python
e2e_infr_dbg_log = open('e2e_inference_dbg_log.txt', 'r')
lines = e2e_infr_dbg_log.readlines()
im2age_map_test_from_ser_e2e_infer = dict()
s = 0
for line in lines:
    if line.startswith('PRINT_TST_IDX_AND_AP_AGE'):
        query_idx = str(int(float(line.split('query_idx=')[1].split(',')[0])))
        query_ap_age = float(line.split('query_ap_age=')[1])
        #print(query_idx, query_ap_age)
        im2age_map_test_from_ser_e2e_infer[query_idx] = query_ap_age
        s += 1
#len(im2age_map_test_from_e2e_infer)
with open("im2age_map_test_from_ser_e2e_infer.json", "w") as outfile:
    json.dump(im2age_map_test_from_ser_e2e_infer, outfile)
```

```python
with open("im2age_map_test.json", 'r') as im2age_map_test_f:
    im2age_map_test = json.load(im2age_map_test_f)
```

```python
d = []
for i in range(len(im2age_map_test_from_ser_e2e_infer)):
    d.append(im2age_map_test_from_ser_e2e_infer[str(i)] - im2age_map_test[str(i)])
```

```python
import matplotlib.pyplot as plt
plt.plot(d)
print(np.mean(np.abs(d)))
```
